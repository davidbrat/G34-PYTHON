```python
#Question 1 (NumPy) – Titanic Dataset Survival Analysis 1. Load the Titanic dataset using Pandas and convert the following numeric columns into a NumPy array: o age, fare, and survived.

Explanation:
We load the Titanic dataset using Pandas with pd.read_csv(), which reads the CSV file into a DataFrame. 
Then, we convert the numeric columns age, fare, and survived into NumPy arrays using the .to_numpy() method for further analysis.
```

```python


import pandas as pd
import numpy as np
df = pd.read_csv(r'Titanic-Dataset.csv')
print(df.head(5))


```

```python
age_array = df['Age'].to_numpy()
print(age_array)
```

```python

fare_array = df['Fare'].to_numpy()

print(fare_array)
```

```python

survived_array = df['Survived'].to_numpy()

print(survived_array)
```

```python
#part-2:Using NumPy (not Pandas), perform the following:
o a) Remove rows with missing values in any of the selected columns.
o b) Normalize age and fare columns using Z-score normalization.
o c) Compute the mean age and mean fare for:
▪ passengers who survived
▪ passengers who did not survive
o d) Use np.where() to classify fare as:
▪ "Low" if fare < mean,
▪ "High" otherwise.
```

```python
a) Removed rows with missing values using ~np.isnan() for clean data.
b) Applied Z-score normalization: (x - mean) / std for age and fare.
c) Used boolean indexing to find mean age and fare for survived and not survived groups.
d) Used np.where() to classify fare as "Low" or "High" based on mean fare.
```

```python
import numpy as np
#a
valid = ~np.isnan(age_array) & ~np.isnan(fare_array) & ~np.isnan(survived_array)

age_clean = age_array[valid]
fare_clean = fare_array[valid]
survived_clean = survived_array[valid]
```

```python
print(age_clean)
```

```python

print(fare_clean)
```

```python

print(survived_clean)
```

```python
#b
import numpy as np
age_mean = np.mean(age_clean)
age_std = np.std(age_clean)
age_normalized = (age_clean - age_mean) / age_std


print("First 10 Normalized Age values:")
print(age_normalized[:10])
```

```python
#b
import numpy as np
fare_mean = np.mean(fare_clean)
fare_std = np.std(fare_clean)
fare_normalized = (fare_clean - fare_mean) / fare_std

print("First 10 Normalized Fare values:")
print(fare_normalized[:10])
```

```python
#c
import numpy as np
survived = survived_clean == 1
not_survived = survived_clean == 0

mean_age_survived = np.mean(age_clean[survived])
mean_fare_survived = np.mean(fare_clean[survived])

mean_age_not_survived = np.mean(age_clean[not_survived])
mean_fare_not_survived = np.mean(fare_clean[not_survived])

print("\n Mean Values by Survival Status :")
print(f"Mean Age (Survived): {mean_age_survived:.2f}")
print(f"Mean Fare (Survived): {mean_fare_survived:.2f}")
print(f"Mean Age (Not Survived): {mean_age_not_survived:.2f}")
print(f"Mean Fare (Not Survived): {mean_fare_not_survived:.2f}")
```

```python
#d
import numpy as np
fare_class = np.where(fare_clean < fare_mean, "Low", "High")

print("First 10 Fare Classifications:")
print(fare_class[:10])
```

```python
import pandas as pd
df = pd.read_excel(r'iris.xlsx')
print(df)
```

```python
#a
df['petal.ratio'] = df['petal.length'] / df['petal.width']
print(df)
```

```python
print(df.columns)
```

```python
#b
avg_petal_ratio = df.groupby('variety')['petal.ratio'].mean()
print(avg_petal_ratio, "\n")
```

```python
#c
std_sepal_length = df.groupby('variety')['sepal.length'].std()
variety_highest_std = std_sepal_length.idxmax()
print(f"{variety_highest_std} ({std_sepal_length.max():.2f})\n")
```

```python

mean_sepal_width = df['sepal.width'].mean()
filtered_df = df[df['sepal.width'] > mean_sepal_width]
print(filtered_df.head(), "\n")
```

```python
#d
subset_low = df[df['petal.ratio'] < 2]
subset_high = df[df['petal.ratio'] >= 2]
combined_df = pd.concat([subset_low, subset_high])
print(combined_df.head())
```
